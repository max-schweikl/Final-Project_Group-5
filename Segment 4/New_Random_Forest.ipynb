```python
df['sucess'] = df.sucess.map(dict(t=1, f=0))
```

```python
pd.set_option('display.max_rows',None)

df.isna().sum()
```

```python
df.fillna(0, inplace=True)
```

```python
X = df.drop(['movie_id', 'index', 'imdb_id', 'original_language', 'original_title', 'popularity', 'release_date', 'title',
            'vote_count', 'sucess_level', 'sucess', 'revenue', 'actor_top100_Other', 'composer_top50_Other', 'director_top50_Other',
            'writer_top50_Other', 'producer_top50_Other'], axis =1 )
y = df['sucess']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=76)
```

```python
# Creating a StandardScaler instance.
# scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
# X_scaler = scaler.fit(X_train)

# Scaling the data.
# X_train_scaled = X_scaler.transform(X_train)
# X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=64, random_state=76) 
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train, y_train)
```

```python
# Making predictions using the testing data.
y_pred = rf_model.predict(X_test)
```

```python
# Calculating the confusion matrix.
cm = confusion_matrix(y_test, y_pred)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm)

cm_df
```

```python
# Calculating the accuracy score.
acc_score = accuracy_score(y_test, y_pred)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, y_pred))
```

```python
# Calculate feature importance in the Random Forest model.
importances = rf_model.feature_importances_
importances
```
